## Naloge

Naloge bodo iz analize smučarskih skokov. Vse funkcije bodo kot argument dobile numpyjevo tabelo (*array*) z dolžinami skokov. Ena funkcija (`stevilo_zmag`) bo dobila dve tabeli. Napišite naslednje funkcije:

Ogrevanje: `povprecje(skoki)` povprečno dolžino skoka. Ta vam je popolnoma podarjena: samo pokličite `np.mean`, ki vrne povprečno vrednost.

### `stevilo_nadpovprecnih`

1. `stevilo_nadpovprecnih(skoki)` vrne število skokov, ki so daljši od povprečne dolžine skoka.

    `stevilo_nadpovprecnih(np.array([200, 100, 110, 210, 212, 215]))` vrne 4, ker so štirje skoki (200, 210, 212, 215) daljši od povprečja (174.5).

#### Rešitev

Če imamo

````python
import numpy as np

skoki = np.array([200, 100, 110, 210, 212, 215])
````

potem je

````python
np.mean(skoki)
````

povprečna dolžina skoka in

````python
skoki > np.mean(skoki)
````

vsebuje `True` na mestih, pri katerih je dolžina skoka daljša od povprečja.

`True`-je preštejemo kar s `sum`, saj je `True` toliko kot 1 in `False` je 0.

````python
np.sum(skoki > np.mean(skoki))
````

Rešitev je torej:

````python
def stevilo_nadpovprecnih(skoki):
    return np.sum(skoki > np.mean(skoki))
````

### `nadpovprecni`

2. `nadpovprecni(skoki)` vrne tabelo dolžin nadpovprečnih skokov.

	`nadpovprecni(np.array([200, 100, 110, 210, 212, 215]))` vrne `np.array([200, 210, 212, 215])`.

#### Rešitev

Nadaljujemo iz tabele `True`-jev in `False`-ov. Uporabimo jo kot masko - z njo indeksiramo tabelo skoki, pa dobimo tabelo tistih, pri katerih imamo vrednost `True`.

````python
skoki[skoki > np.mean(skoki)]
````

Rešitev je torej

````python
def nadpovprecni(skoki):
    return skoki[skoki > np.mean(skoki)]
````

### `povprecje_nadpovprecnih`

3. `povprecje_nadpovprecnih(skoki)` vrne povprečno dolžino nadpovprečnih skokov.

	`povprecje_nadpovprecnih(np.array([200, 100, 110, 210, 212, 215]))` vrne 209.25, ker je povprečje skokov (200, 210, 212, 215) enako 209.25.

#### Rešitev

Izračunamo samo povprečje tega, kar vrne prejšnja funkcija.

````python
def povprecje_nadpovprecnih(skoki):
    return np.mean(nadpovprecni(skoki))
````

### `stevilo_zmag`

4. `stevilo_zmag(skoki, drugi)` primerja dva tekmovalca, katerih skoki so podani v tabelah `skoki` in `drugi`. Vrne število skokov, kjer je prvi tekmovalec skočil dlje od drugega. Primerjamo istoležne skoke (prvi skok prvega tekmovalca z prvim skokom drugega tekmovalca, drugi z drugim itd.).

	```
	stevilo_zmag(np.array([200, 100, 110, 210, 212, 215]),
	             np.array([190, 120, 130, 200, 220, 210]))
	```
	
	vrne 4, ker je prvi tekmovalec skočil dlje v skokih prvem in v zadnjih treh skokih.

#### Rešitev

Primerjajmo tabeli, pa dobimo tabelo, ki vsebuje `True` na mestih, kjer je prvi daljši in `False` tam, kjer ni.

````python
prvi = np.array([200, 100, 110, 210, 212, 215])
drugi = np.array([190, 120, 130, 200, 220, 210])

prvi > drugi
````

Preštejemo `True`-je.

````python
np.sum(prvi > drugi)
````

````python
def stevilo_zmag(skoki, drugi):
    return np.sum(skoki > drugi)
````

### `vedno_boljsi`

5. `vedno_boljsi(skoki)` vrne `True`, če je vsak naslednji skok daljši od prejšnjega, sicer `False`.

	`vedno_boljsi(np.array([100, 150, 200, 250]))` vrne `True`, ker je vsak naslednji skok daljši od prejšnjega.

	`vedno_boljsi(np.array([100, 150, 150, 250]))` vrne `False`, ker drugi in tretji skok nista v naraščajočem vrstnem redu.

	`vedno_boljsi(np.array([100, 80, 150]))` vrne `False`, ker je drugi skok krajši od prvega.

#### Rešitev

Tu primerjamo prvi element z ničtim, drugega s prvim in tako naprej.

````python
skoki[1:] > skoki[:-1]
````

Ali so v tabeli sami `True`, nam pove `np.all`.

````python
def vedno_boljsi(skoki):
    return np.all(skoki[1:] > skoki[:-1])
````

Tule so nekateri študenti odkrili `diff`.

````python
np.diff(skoki)
````

`diff` izračuna razlike med zaporednimi elementi. Zanima nas, ali so vse razlike pozitivne, torej

````python
np.diff(skoki) > 0
````

... potem pa spet `np.all`.

Da, lahko tudi tako. `np.diff` je sicer močnejša zverina, ki zmore še marsikaj zahtevnejšega od razlik od zaporednih elementov (po eni osi).

### `naj_skoki`

6. `naj_skoki(skoki)` vrne zaporedne številke najdaljših skokov.

	`naj_skoki(np.array([100, 200, 150, 200, 50]))` vrne `np.array([2, 4])`, ker sta najdaljša 2. in 4. skok (oba sta dolga 200 m).

#### Rešitev

Najprej z `np.max` izvemo dolžino najdaljšega skoka.

````python
skoki = np.array([100, 200, 150, 200, 50])

np.max(skoki)
````

Potem poizvemo, kje v tabeli se pojavi ta številka.

````python
skoki == np.max(skoki)
````

Indekse `True`-jev nam pove `np.flatnonzero`.

````python
np.flatnonzero(skoki == np.max(skoki))
````

Naloga hoče, da štejemo od 1, ne od 0, zato k tem indeksom prištejemo 1.

````python
np.flatnonzero(skoki == np.max(skoki)) + 1
````

````python
def naj_skoki(skoki):
    return np.flatnonzero(skoki == np.max(skoki)) + 1
````

Kar nekaj študentov je namesto `flatnonzero` uporabilo `where`. To deluje, vendar je `where` namenjen drugim stvarem in tudi dokumentacija numpyja priporoča, da tule uporabimo `nonzero` oziroma, v našem primeru, preprostejši `flatnonzero`.

### `ocena`

7. Tale je malo bolj zanimiva (nič težkega, samo malo morate pomisliti): `ocena(skoki)` vrne oceno, ki jo izračunamo tako, da izločimo najkrajši in najdaljši skok ter izračunamo povprečno dolžino ostalih. (Nasvet: pri tem nisem uporabil funkcije `mean`, pa tudi urejal nisem.)

	`ocena(np.array([100, 200, 150, 300, 250]))` vrne 200, ker je povprečje skokov (200, 150, 250) enako 200.

#### Rešitev

Povprečje je enako vsoti, deljeni s številom elementov. Če pri tem ne želimo upoštevati največjega in najmanjšega elementa, ju pač odštejemo od vsote.

````python
def ocena(skoki):
    return (np.sum(skoki) - np.min(skoki) - np.max(skoki)) / (len(skoki) - 2)
````
