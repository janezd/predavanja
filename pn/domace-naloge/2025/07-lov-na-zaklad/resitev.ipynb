## Skrivna sporočila

Prebrati morate pet skrivnih sporočil. Za vsako sporočilo vemo, kje se začne. Prvo, recimo, se začne v direktoriju `zelje-testenine/luna-avtobus/blitva-drevo` (znotraj poddirektorija `datoteke` - to morate sami dodati na začetek!), številka vrstice pa je 42 (vrstice bomo šteli od 0, ne od 1!).

V tem direktoriju je nekaj poddirektorijev in ena sama datoteka s končnico .txt (slučajno se imenuje `otrok-lopar.txt`). Če odpremo to datoteko in preberemo vrstico 42 (ker bomo vrstice šteli od 0, urejevalniki kot je PyCharm, če datoteko odpremo v njem, pa štejejo od 1, to vrstico tam vidimo kot 43. vrstico), najdemo:

```
N,reka-med/marelica-zvezda/grozdje-grozdje/kraljica-enorog,25
```

To pomeni, da je prva črka skrivnega sporočila N, naslednje sporočilo pa najdemo v direktoriju `reka-med/marelica-zvezda/grozdje-grozdje/kraljica-enorog`, zanimala nas bo vrstica 25.

V tem direktoriju poiščemo datoteko s končnico .txt. Imenuje se `enorog-teta.txt` in v 25. vrstici (26. v urejevalniku) najdemo

```
A,boj-princesa/trg-sosed/brokoli-rama,79
```

Drugi znak skrivnega sporočila je torej A, naslednje sporočilo pa najdemo v direktoriju `boj-princesa/trg-sosed/brokoli-rama`, vrstica 79.

Tako nadaljujemo, dokler ne pridemo do pike.

- Med znaki bomo našli tudi presledke. V tem primeru bo imela vrstica pred (prvo) vejico preseledek. (Torej: ne kličite `strip()`!)
- V vsakem direktoriju je natanko ena datoteka s končnico .txt. Njeno ime je naključno in nepomembno.

## Rešitev

#### Pomožna funkcija

Ubogajmo, napišimo si pomožno funkcijo. Poimenujmo je `preberi_vrstico`, kot argument pa bo dobila ime direktorija (na primer `"boj-princesa/trg-sosed"` in številko vrstice, ki nas zanima.

````python
def preberi_vrstico(direktorij, st_vrstice):
    for ime in os.listdir(os.path.join("datoteke", direktorij)):
        if ime.endswith(".txt"):
            f = open(os.path.join("datoteke", direktorij, ime), encoding="utf-8")
            for _ in range(st_vrstice):
                f.readline()
            vrstica = f.readline()
            crka, novi_direktorij, nova_stevilka = vrstica.split(",")
            return crka, novi_direktorij, int(nova_stevilka)
````

Ker argument funkcije ne vsebuje `datoteka`, temveč se začne šele z bojem in princeso, moramo na začetek imena direktorija prilepiti še `datoteke`. To lahko storimo z `"datoteke/" + direktorij` ali `f"datoteke/{direktorij}" (če to že znamo) ali `os.path.join("datoteke", direktorij)`. Najbolj ugledna je zadnja rešitev.

Za vsako ime, ki ga dobimo od `os.listdir` preverimo, ali se konča s ".txt". To spet lahko storimo na več načinov: trije med njimi so `ime.endswith(".txt")`, `ime[-4:] == ".txt"` in `os.path.splitext(ime)[1] == ".txt"`. Prvi je najpreprostejši, drugega uporabimo, če se ne spomnimo na `endswith`, tretji pa je, recimo "najpravilnejši", vendar me moti tisti `[1]`, pri katerem se lahko zmotiš in napišeš tudi `[0]`. (Paranoja? Ne, temu se reče *modrost*. "Vse to sem že videl, vse že doživel," je zapel oni, ki mu je šla s prsti skozi lase.)

Nato preskočimo prvih `st_vrstice` vrstic; če je `st_vrstice` enak 3, moramo preskočiti ničto, prvo in drugo, torej ravno tri. Nato preberemo tisto, ki nas zanima.

Konec je preprost: vrstico razbijemo glede na vejico, dobljene dele poimenujemo `crka`, `novi_direktorij` in `nova_stevila`, pri čemer slednjo pretvorimo v `int`, da bo res številka.

Ne brez elegance je malo drugačna rešitev:

````python
def preberi_vrstico(direktorij, st_vrstice):
    for ime in os.listdir(os.path.join("datoteke", direktorij)):
        if ime.endswith(".txt"):
            break
    for i, vrstica in enumerate(open(os.path.join("datoteke", direktorij, ime), encoding="utf-8")):
        if i == st_vrstice:
            break
    crka, novi_direktorij, nova_stevilka = vrstica.split(",")
    return crka, novi_direktorij, int(nova_stevilka)
````

Zanki sta krajši, razložljivejši, pa še neko privlačno simetrijo smo dobili. Prva zanka jemlje imena, dokler ne naleti na takšno, ki se konča s `.txt`; po tem, je `ime` ime datoteke, ki jo želimo brati. Druga zanka bere to datoteko, dokler ne pride do vrstice, ki nas zanima. Konec je pa tak kot prej.

#### Glavni del

Funkcija `beri` je zdaj kratka.

````python
def beri(direktorij, vrstica):
    sporocilo = ""
    crka = None
    while crka != ".":
        crka, direktorij, vrstica = preberi_vrstico(direktorij, vrstica)
        sporocilo += crka
    return sporocilo
````

Skoraj ni kaj komentirati: v `sporocilo` beremo črke sporočila, dokler ne dodamo pike, po čemer zanko ustavimo.

Zanimiva alternativa je, da preverjamo, da zadnja črka sporočila ni pika. Pri tem ne smemo uporabiti `sporocilo[-1]`, saj je `sporocilo` v začetku prazen niz. Pač pa smemo varno napisati `sporocila[-1:]`. Rezine, npr. `s[:n]` v primeru, da ima `s` manj kot `n` elementov, pač vrnejo celotni `s`. Tu bomo torej dobili zadnjo črko (ki bo ali pa ne bo pika), ko bo niz prazen, pa bomo s `sporocilo[-1]` dobili prazen niz in pogoj `sporocilo[-1] == "."` bo neresničen.

````python
def beri(direktorij, vrstica):
    sporocilo = ""
    while sporocilo[-1:] != ".":
        crka, direktorij, vrstica = preberi_vrstico(direktorij, vrstica)
        sporocilo += crka
    return sporocilo
````

Nek študent je nalogo rešil z rekurzivno funkcijo. Ta rešitev mi je sicer prišla na misel ob sestavljanju naloge, vendar sem jo odmislil, ker ta problem *ne zahteva* rekurzije. (Ta izjava je nekoliko nevarna. Noben problem ne zahteva rekurzije, pač pa je nekatere stvari bolj naravno reševati rekurzivno in tole na prvi pogled ni ena teh stvari.) Kakorkoli, izkaže se, da je rekurzivna rešitev prav lepa in vredna biti napisana.

````python
def beri(direktorij, vrstica):
    crka, direktorij, vrstica = preberi_vrstico(direktorij, vrstica)
    if crka == ".":
        return crka
    return crka + beri(direktorij, vrstica)
````

Če preberemo piko, vrnemo piko, sicer pa črko, ki smo jo dobili in vse črke, ki jih preberemo kasneje (z isto funkcijo).
