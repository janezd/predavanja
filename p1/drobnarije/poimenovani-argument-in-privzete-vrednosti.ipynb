## Poimenovani argumenti

Ko pokličemo funkcije, v oklepaju podamo vrednosti argumentov.

````python
pow(5, 3)
````

Včasih pa smo argument podali kot nekakšno prirejanje, recimo pri odpiranju datoteke. 

````python
open("stevilke.txt", encoding="utf-8")
````

Zakaj `encoding="utf-8"` in ne preprosto `"utf-8"`?

Funkcija `open` ima več argumentov in tule eksplicitno povemo, vrednost katerega argumenta določamo.

Napišimo preprosto - in ne nujno najbolj uporabno - funkcijo: sprejela bo štiri argumente in iz njih sestavila število.

````python
def stevilka(enice, desetice, stotice, tisocice):
    return enice + 10 * desetice + 100 * stotice + 1000 * tisocice
````

````python
stevilka(4, 1, 7, 9)
````

Funkcijo lahko pokličemo tudi tako, da vse - ali nekatere - argumente poimenujemo.

````python
stevilka(enice=4, desetice=1, stotice=7, tisocice=9)
````

````python
stevilka(4, 1, stotice=7, tisocice=9)
````

Lepota poimenovanih argumentov je v tem, da nam ni treba poznati vrstnega reda v definiciji funkciji. Navedemo jih lahko v poljubnem vrstnem redu - seveda, dokler nobenega ne pozabimo.

````python
stevilka(stotice=7, tisocice=9, desetice=1, enice=4)
````

````python
stevilka(4, 1, tisocice=9, stotice=7)
````

````python
stevilka(stotice=7, desetice=1)
````

````python
stevilka(4, stotice=7, desetice=1)
````

To še ni zelo uporabno. Pač pa bi nam prišlo prav, če bi lahko podali le nekatere argumente, ostali pa bi imeli privzete vrednosti.

## Argumenti s privzetimi vrednostmi

V definiciji funkcije lahko določimo privzete vrednosti argumentov - vseh, ali pa vseh od določenega argumenta naprej.

````python
def stevilka(enice=0, desetice=0, stotice=0, tisocice=0):
    return enice + 10 * desetice + 100 * stotice + 1000 * tisocice
````

Če zdaj kak argument izpustimo, bo imel vrednost `0`. V splošnem lahko, seveda, določimo tudi kakšno drugo privzeto vrednost; tu je pač najbolj smiselna 0.

````python
stevilka(2, 4)
````

````python
stevilka(4)
````

````python
stevilka()
````

Zdaj pa to skombinirajmo s prejšnjim: poimenovanimi argumenti.

````python
stevilka(tisocice=2, desetice=2, enice=6)
````

Srečno novo leto!

Funkcija `open` ima, tako kot mnoge Pythonove funkcije, privzete vrednosti argumentov. Funkcija `open`, recimo, takšne:

```
open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
```

Prvi argument nima privzete vrednosti: `open` potrebuje ime datoteke. Naslednji argument, `mode`, pove, ali bomo datoteko brali, pisali ali dopolnjevali in ali bo besedilna ali binarna. Tudi ta argument smo vedno podajali brez imena. Tako *vedno*, da se vaš predavatelj Programiranja 1 na predavanju niti ne bi spomnil imena tega argumenta. Prav tako vaš predavatelj med predavanjem praktično ni vedel za argument `buffering`, s katerim lahko določimo velikost pomnilnika, v katerega Python shranjuje, kar zapišemo v datoteko z `write`, še preden se to dejansko shrani na disk. (Pravzaprav ne: preden Python to preda v pisanje operacijskemu sistemu, ki pa spet zavlačuje s pisanjem, dokler more.) Tudi ostale argumente vidim takorekoč prvič - točneje, vsakič, ko jih vidim, je, kot da bi jih videl prvič.

Kar dokazuje, kako uporabni so argumenti s privzetimi vrednostmi in poimenovani argumenti: omogočijo nam, da uporabljamo le tisto, kar nas zanima, ostalo pa pustimo pri miru.
