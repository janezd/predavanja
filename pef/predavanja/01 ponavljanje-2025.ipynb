## 1. Robotski sesalec

Imamo robotski sesalec, ki ima štiri osnovne smeri gibanja (brez obračanja in orientacije) ter razume zgolj binarne vrednosti `0` in `1`. Ukazi so zapisani kot niz znakov `'0'` in `'1'` sode dolžine. Robot bere niz po dva znaka naenkrat, pri čemer vsak par predstavlja en ukaz. Robot začne v točki `(0, 0)`.

- `00` - pomeni premik gor
- `01` - pomeni premik dol
- `10` - pomeni premik levo
- `11` - pomeni premik desno

Napišite funkcijo robot(navodilo), ki vrne končne koordinate robota po izvedbi vseh ukazov.

```
>>> robot("001001")
(-1, 0)
>>> robot("10111100")
(1, 1)
```

### Rešitev malo drugačne naloge

Tudi sam pogosto dajem naloge, v katerih je potrebno slediti gibanju nekega objekta ali subjekta (pogosto: kolesarja), pri čemer pa smeri podajam malo drugače. V moji nalogi bi moral robot sprejemati nize v slogu `"<^^>>v"`; pomeni znakov so očitni. Rešimo najprej takšno različico naloge.

````python
def robot(navodilo):
    x = y = 0
    for smer in navodilo:
        if smer == "^":
            y += 1
        elif smer == "v":
            y -= 1
        elif smer == "<":
            x -= 1
        elif smer == ">":
            x += 1
    return (x, y)
````

````python
robot("<^^>>v")
````

Tule lahko takoj napišemo milijon komentarjev. No, ali kakšnega manj.

- Sam sem nalogo najprej rešil narobe: predpostavil sem, da `y` narašča v smeri navzdol. To je zato, ker pred karirastim papirjem razmišljam kot matematik, za računalnikom po kot računalnikar. V računalniški grafiki `y` narašča navzdol. Če se vam to zdi čudno, poiščite sedmo vrstico (prvega poglavja prve knjige o Harryju Potterju). Našli? OK, ste šteli od zgoraj navzdol ali od spodaj navzgor. Sorry, čudni so matematiki, ne računalnikarji.
- Za prirejanje začetnih vrednosti `x` in `y` imamo kup variant.
  1. `x = y = 0`: `x` in `y` nastavi na `0`.
  2. `x = 0` in v naslednji vrstici `y = 0`: nastavi `x` na `0` in nato nastavi `y` na 0.
  3. `x, y = 0, 0`: `x` in `y` nastavi na `0` in `0`.
  4. `x, y = (0, 0)`: `x, y` nastavi na koordinate `(0, 0)`.
   
  Ko smo nekje v Bosni v osemdesetih naročali eni dunajski, eni pariški in eni naravni zrezek, je natakar izrekel nepozabne besede "*pa sve je to isto!*". Varianta 1 je najkrajša in tehnično čisto OK; Pythonova priporočila dopuščajo takšno veriženje, kadar spremenljivke predstavljajo "podobne reči". Sicer je bolj priporočena druga možnost, ki pa je tu "semantično" manj zgovorna. Tretja je OK, četrta pa najbolj jasno pove, kaj počnemo.

  Tole je seveda dlakocepstvo; povedati hočem le, da bomo pri programiranju pogosto razmišljali o tem, kaj v bistvu hočemo povedati in kako temu, ki bere program, čimbolj olajšati branje.

O oklepajih v `return`-u sem se razpisal, ker jih velikokrat videvam pri študentih. Navadno se izkaže, da jih niso pisali sami, temveč jim je pri reševanju domače naloge "pomagal" nekdo, ki je teh oklepajev vajen iz kakega drugega jezika. Pa to ni v čast ne študentu ne "pomočniku".

### batched

## Končno rešitev izpitne naloge

Zato bi uporabil drugi način, kadar delam na projektu, v katere smem uporabljati Python 3.12 (počasi bodo to vsi projekti, na katerih delam), pa je očitno najbolj praktično uporabiti `batched`. Rešitev bo torej:

````python
def robot(navodilo):
    x = y = 0
    for d, e in batched(navodilo, 2):
        smer = d + e
        if smer == "00":
            y += 1
        elif smer == "01":
            y -= 1
        elif smer == "10":
            x -= 1
        elif smer == "11":
            x += 1
    return x, y
````

### Bolj zabavne rešitve

`e` je skoraj `int` (`"0"` ali `"1"` lahko z enim klicem funkcije `int` spremenimo v `0` in `1`), `int`-i pa lahko služijo kot indeksi, ne?

````python
def robot(navodilo):
    x = y = 0
    for d, e in batched(navodilo, 2):
        diff = [1, -1][int(e)]
        if d == "0":
            y += diff
        else:
            x += diff
    return -x, y
````

````python
robot(s)
````

Torej: `int(e)` je `0` ali `1`. Če je `e` enak `0`, hočemo, da je `diff` `1`, če je `e` `1`, pa mora biti `diff` `-1`. Pripravimo torej seznam `[1, -1]` in vzamemo element, z indeksom `int(e)`.

Zdaj pa bi naredili podogen trik še z `d`. Namesto spremenljivk `x` in `y`, pripravimo seznam, ki bo vseboval koordinati; ničti element bo koordinata `y`, prvi pa `x`. V vsakem koraku zanke spremenimo `int(e)`-ti element.

````python
def robot(navodilo):
    coord = [0, 0]
    for d, e in batched(navodilo, 2):
        diff = [1, -1][int(e)]
        coord[int(d)] += diff
    return -coord[1], coord[0]
````

````python
robot(s)
````

Ali, z malo manj dolgovezenja:

````python
def robot(navodilo):
    coord = [0, 0]
    for d, e in batched(navodilo, 2):
        coord[int(d)] += [1, -1][int(e)]
    return -coord[1], coord[0]
````

````python
robot(s)
````

````python
def robot(navodilo):
    s = list(batched(navodilo, 2))
    return s.count(("1", "1")) - s.count(("1", "0")), s.count(("0", "0")) - s.count(("0", "1")) 
````

## 1.5 Poldruga naloga: Posesana površina

Kvaliteta sesalca kajpa ni odvisna od tega, kako daleč se odpelje, temveč od tega, koliko polj posesa. Rešimo malo drugačno nalogo: napišimo funkcijo `povrsina(navodila)`, ki vrne število polj, ki jih obišče sesalec. Istega polja seveda ne štejemo večkrat. Sesalec je dober in polje brezhibno posesa že ob prvem prehodu.

Z našim znanjem lahko to storimo le tako, da obiskana polja shranjujemo v seznam -- pri čemer polja, ki je že obiskano, seveda ne dodajamo. Da ne bo prezapleteno, začnimo s takšnole različico rešitve gornje naloge:

````python
def robot(navodilo):
    x = y = 0
    for i in range(0, len(navodilo), 2):
        smer = navodilo[i:i + 2]
        print(smer)
        if smer == "00":
            y += 1
        elif smer == "01":
            y -= 1
        elif smer == "10":
            x -= 1
        elif smer == "11":
            x += 1
    return x, y
````

````python
robot(s)
````

Zdaj pa dodajmo seznam obiskanih polj.

````python
def povrsina(navodilo):
    x = y = 0
    obiskana = [(0, 0)]
    for i in range(0, len(navodilo), 2):
        smer = navodilo[i:i + 2]
        if smer == "00":
            y += 1
        elif smer == "01":
            y -= 1
        elif smer == "10":
            x -= 1
        elif smer == "11":
            x += 1
        if (x, y) not in obiskana:
            obiskana.append((x, y))
    return len(obiskana)
````

````python
povrsina(s)
````

## 2. Mapiranje

Robotski sesalec iz 1. naloge uporabljamo v sobi pravokotne oblike velikosti širina × višina. Soba je razdeljena na polja, katerih koordinate so `(x, y)`, kjer velja 0 ≤ x < širina in 0 ≤ y < višina. Sesalec začne v `(0, 0)` – spodnji levi kot. Ko je sesalec na polju, ga avtomatično posesa.

Napišite funkcijo `mapiraj(sirina, visina)`, ki vrne niz `'0'` in `'1'` (dolžine sodo število), ki predstavlja zaporedje ukazov (`00` gor, `01` dol, `10` levo, `11` desno) tako, da sesalec med izvajanjem ukazov obišče vsako polje sobe vsaj enkrat. Sesalec se ne sme premakniti izven sobe.

### Učinkovitejši sesalec

Predlagana pot mojega sesalca je bistveno daljša od poti, ki jo prevozi sesalec doc. Zapuška, `">>^<<"` (v izpitni nalogi `"1111001010"`). Ker mi je kot profesorju v sramoto, da bi docenti boljše programirali sesalce kot jaz, bomo morali premagati lenobo in rešiti nalogo še na ta način. (**Disclaimer**: "prof." Demšar je videl že toliko zanikrnih profesorjev in odličnih docentov, da mu dol visi za titule. Še več, prepričan sem, da obstaja kar nekaj razlogov, zaradi katerih je povprečen docent boljši predavatelj od povprečnega profesorja. Eden je starost in motivacija, drugi je, da predani pedagogi pogosto nimajo časa za pisanje člankov, ki so pogoj za višje naslove. Predvsem pa: doc. Zapušek, ki si ga občasno takole privoščujem ob reševanju izpita, je odličen asistent in predavatelj.)

Pa naredimo še sesalec, ki gre gor, potem pa dol v naslednjem "stolpcu". Saj ni tako težko.

````python
def mapiraj(sirina, visina):
    navodilo = ""
    for i in range(visina):
        if i != 0:
            navodilo += "^"
        if i % 2 == 0:
            navodilo += ">" * (sirina - 1)
        else:
            navodilo += "<" * (sirina - 1)
    return navodilo
````

````python
mapiraj(3, 2)
````

Takole: naredili bomo `visina` sprehodov levo in desno, torej `for i in range(visina)`. Vsako - razen prve, to je, ničte! - bomo začeli s korakom gor, `00`. Nato bomo v sodih vrsticah naredili sirina - 1 korakov desno (11), v lihih pa natančno toliko korakov levo (10).

Spet se lahko malo poigramo z indeksi. Ker je zabavno.

````python
def mapiraj(sirina, visina):
    navodilo = ""
    for i in range(visina):
        if i != 0:
            navodilo += "^"
        navodilo += [">", "<"][i % 2] * (sirina - 1)
    return navodilo
````

````python
mapiraj(3, 2)
````

Na razpolago sta dva premika, desno in levo, `>` in `<`. Zložimo ju v seznam in ostanek po deljenju `i` z `2` pove, katerega moramo vzeti.

Mimogrede, z našimi znaki bi šlo tudi brez seznama.

````python
def mapiraj(sirina, visina):
    navodilo = ""
    for i in range(visina):
        if i != 0:
            navodilo += "^"
        navodilo += "><"[i % 2] * (sirina - 1)
    return navodilo
````

````python
mapiraj(3, 2)
````

Pa še, mimogrede, če hočemo ničle in enke, lahko vseeno programiramo z mojimi znaki in jih na koncu funkcije ustrezno zamenjamo.

````python
def mapiraj(sirina, visina):
    navodilo = ""
    for i in range(visina):
        if i != 0:
            navodilo += "^"
        navodilo += "><"[i % 2] * (sirina - 1)
    return navodilo.replace(">", "11").replace("<", "10").replace("^", "00")
````

````python
mapiraj(3, 2)
````

Znebimo se lahko tudi prvega `if`, vendar to za zdaj preskočimo.

Bolj bo zabavno, če nalogo rešimo kar brez zanke, takole.

````python
# Tole ni pravilna rešitev!
def mapiraj(sirina, visina):
    return (">" * (sirina - 1) + "^" + "<" * (sirina - 1) + "^") * (visina // 2)
````

````python
mapiraj(3, 2)
````

Najprej poglejmo, zakaj to deluje, in potem, zakaj ne (ker je očitno že tule dalo napačen rezultat).

V prvem oklepaju gremo `sirina - 1`-krat desno, nato gor in nato `sirina - 1`-krat levo in gor. Znotraj oklepaja torej posesamo dve vrstici. Zato to ponovimo `visina // 2`-krat. (Saj se spomnimo? `//` pomeni celoštevilsko deljenje. Deljenje je zaokroženo navzdol in rezultat je `int`.)

Tri težave imamo.

- Prva, očitna je `^` na koncu. A to znamo rešiti, to smo že počeli.
- Druga je, da je deljenje zaokroženo navzdol. Če je število vrstic liho, bo sesalec pozabil posesati zadnjo.
- Tretje še nimamo, dobili pa jo bomo, čim bomo deljenje zaokrožili navzgor, saj bo sesalec posesal še eno odvečno vrstico.

Začnimo pri drugi težavi.

````python
mapiraj(3, 3)
````

Da, tole je posesalo le dve vrstici - rezultat je enak kot pri `mapiraj(3, 2)`.

Za zaokrožanje celoštevilskega deljanja navzgor obstaja lep trik: prištejemo 1.

````python
x = 5
````

````python
(x + 1) // 2
````

Trik je splošnejši: če delimo z `n`, moramo k deljenjcu prišteti `n - 1`, pa bo deljenje zaokroženo navzgor. Če 50 delimo s 5, bomo prišteli 4 in 54 // 5 bo še vedno 10. Če je bilo število prej deljivo z `n`, prištevanje `n - 1` še ne bo spremenilo količnika. Čim je število za 1 večje, pa bo že dovolj za zaokrožanje navzgor: če 51 delimo s 5, bomo prišteli 4 in 55 / 5 je že 11.

Spremenimo torej svojo funkcijo tako, da zaokrožimo `visina // 2` navzgor.

````python
# Tole ni pravilna rešitev!
def mapiraj(sirina, visina):
    return (">" * (sirina - 1) + "^" + "<" * (sirina - 1) + "^") * ((visina + 1) // 2)
````

Če je višina soda, se seveda nič ne spremeni.

````python
mapiraj(3, 2)
````

Če je liho, pa posesamo eno vrstico preveč.

````python
mapiraj(3, 3)
````

Tole očitno posesa štiri vrstice - dve v smeri `>` in dve v smeri `<`.

Ostali sta nam torej tretja in prva težava. Rešili bomo kar obe hkrati. Vemo namreč, kako dolg mora biti niz, ki ga vračamo. Za vsako vrstici naredimo `sirina - 1` korakov desno ali levo in 1 korak gor. Skupaj torej ravno `sirina` korakov v vsaki vrstici. Izjema je prva, v kateri ni koraka gor. Vemo torej, da mora imeti niz, ki ga vračamo, `sirina * visina - 1` znakov. Vsi, ki jim, sledijo, so odveč.

````python
def mapiraj(sirina, visina):
    return ((">" * (sirina - 1) + "^" + "<" * (sirina - 1) + "^") * ((visina + 1) // 2))[:sirina * visina - 1]
````

````python
mapiraj(3, 2)
````

````python
mapiraj(3, 3)
````

Če hočemo rešiti originalno nalogo, lahko, tako kot prej, zamenjamo smeri z ničlami in enkami.

````python
def mapiraj(sirina, visina):
    return ((">" * (sirina - 1) + "^" + "<" * (sirina - 1) + "^") * ((visina + 1) // 2))[:sirina * visina - 1].replace(">", "11").replace("<", "10").replace("^", "00")
````

Ker je vrstica nekoliko dolga, jo zlomimo; lahko kar znotraj kakega oklepaja, da bo Python vedel, da se vrstica nadaljuje v naslednji vrstici.

````python
def mapiraj(sirina, visina):
    return ((">" * (sirina - 1) + "^" + "<" * (sirina - 1) + "^") * ((visina + 1) // 2)
           )[:sirina * visina - 1].replace(">", "11").replace("<", "10").replace("^", "00")
````

Ali, da bo prelom lepši, v prvi vrstici dokončamo niz, nato z `\` povemo, da se vrstica nadaljuje in v drugi naredimo še ustrezne zamenjave.

````python
def mapiraj(sirina, visina):
    return ((">" * (sirina - 1) + "^" + "<" * (sirina - 1) + "^") * ((visina + 1) // 2))[:sirina * visina - 1] \
        .replace(">", "11").replace("<", "10").replace("^", "00")
````
