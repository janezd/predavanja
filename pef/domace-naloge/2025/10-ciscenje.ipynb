## Čiščenje

Dajmo eno čisto preprosto: napišite funkcijo `prepisi(vhod, izhod)`, ki dobi imena dveh datotek - očitno vhodne, iz katere bere podatke, in izhodne, v katero bo pisala.

Vhodna vsebuje imena modelov sesalcev, letnico izdelave (23 pomeni 2023) in ceno **v yuanih**, na primer takole:

```
Sesalec,23,284
Čistoses,24,783
Čistoses Deluks,25,1024
Sesalopuhež,25,583
Pušek,26,845
```

V izhodno datoteko mora izpisati tabelo v **natančno takšni obliki**, le podatki bodo seveda lahko drugačni.

```
Model                         Letnik      Cena
----------------------------------------------
Sesalec                         2023     35.50
Čistoses                        2024     97.88
Čistoses Deluks                 2025    128.00
Sesalopuhež                     2025     72.88
Pušek                           2026    105.62
----------------------------------------------
```

- Vhodna datoteka je zapisana kot cp-1250. Izhod mora biti shranjen v UTF-8.
- Cena v izhodni datoteki je v evrih. Predpostaviti smete, da 1 evro velja 8 yuanov.
- Da ne boste šteli: model je izpisan na 30 mest, letnik na 6 mest in cena na 10 mest z dvema decimalkama. Poravnava je očitna.

## Rešitev

````python
def prepisi(vhod, izhod):
    g = open(izhod, "w", encoding="utf-8")
    g.write(f"{'Model':30}{'Letnik':6}{'Cena':>10}\n")
    g.write(f"{'-' * (30 + 6 + 10)}\n")
    for vrstica in open(vhod, encoding="cp1250"):
        tip, letnik, cena = vrstica.split(",")
        g.write(f"{tip:30}{2000 + int(letnik):6}{float(cena) / 8:10.2f}\n")
    g.write(f"{'-' * (30 + 6 + 10)}\n")
````
