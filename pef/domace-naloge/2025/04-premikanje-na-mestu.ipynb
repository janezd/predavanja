## Rešitev

Za prvo nalogo je pomembno, da spreminjamo seznam, ki smo ga prejeli. Tole, recimo, ne deluje:

````python
def premik(koordinate, smer):
    x, y = koordinate
    if smeri == "^":
        y += 1
    elif smeri == "v":
        y -= 1
    elif smeri == "<":
        x -= 1
    elif smeri == ">":
        x += 1
    koordinate = [x, y]  # Napaka!
````

Problem je zadnja vrstica. Ta ne spremeni seznama, ki ga je funkcija dobila kot argument (in na katerega se v začetku nanaša ime `koordinate`), temveč v zadnji vrstici določi, naj se `koordinate` nanaša na nek drug, novi seznam.

Funkcija bi delovala, če bi zadnjo vrstico zamenjali z

```python
    koordinate.clear()
    koordinate += [x, y]
```

saj zdaj spreminjamo seznam, na katerega se nanaša ime `koordinate` (torej seznam, ki smo ga dobili kot argument).

Še običajnejše bi napisali

```python
    koordinate[:] = [x, y]
```

To zamenja vse elemente (`koordinate[:]`) z elementoma `[x, y]`.

Ta rešitev je seveda brez potrebe zapletena. Pri študentih sem videval tole.

````python
def premik(koordinate, smer):
    if smer == "^":
        koordinate[1] += 1
    elif smer == "v":
        koordinate[1] -= 1
    elif smer == ">":
        koordinate[0] += 1
    elif smer == "<":
        koordinate[0] -= 1
````

Jaz pa sem pacek in sem naredil tako:

````python
def premik(koordinate, smer):
    koordinate[smer in "v^"] += 1 - 2 * (smer in "<v")
````

Izraz `smer in "v^"` je resničen (`True`) ali neresničen (`False`). Ker je `True` toliko kot `1`, bo `smer in "v^"` v bistvu enak `1`, če se premikamo v smeri `y` in `0`, če v smeri `x`. Tako torej določimo, katero koordinato bomo spreminjali.

Tudi izraz `smer in "<v"` je resničen ali neresničen. Resničen, torej `True` oz. `1` bo takrat, ko gremo levo oziroma dol, se pravi, ko je potrebno od koordinate odšteti 1. Ali, drugače, ko je potrebno h koordinati *prišteti `-1`*. `1 - 2 * (smer in "<v")` bo enako `-1`, če je oni izraz resničen in `1`, če ni.

Trik druge naloge je, da moramo pripraviti kopijo seznama, preden kličemo premik.

Tole ne deluje:

````python
def premik_ali_ne(koordinate, smer):
    prej = koordinate  # Napaka
    premik(koordinate, smer)
    return prej != koordinate
````

Tokrat je napaka v prvi vrstici, ki ne pripravi novega seznama temveč le novo ime za obstoječega. Ko `premik` spremeni (ali pa ne spremeni) `koordinate`, tako spreminja tudi `prej`. Ker gre za *isti* seznam, bo pogoj `prej != koordinate` vedno neresničen; seznama sta ista, torej enaka.

Pravilna rešitev je

````python
def premik_ali_ne(koordinate, smer):
    prej = koordinate.copy()
    premik(koordinate, smer)
    return prej != koordinate
````

Namesto `koordinate.copy()` pogosto pišemo `koordinate[:]`. Ko gre za sezname, je vseeno. Če hočemo narediti kopijo slovarja ali množice, pa moramo poklicati `copy()`, saj slovarji in množice ne poznajo rezin.
