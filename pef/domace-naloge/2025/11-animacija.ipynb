## Drugi del

Zdaj pa imamo opravka s sesalcem, ki navodila intepretira malo drugače. Sesalec je v začetku levo spodaj in je obrnjen na desno.

- `^` pomeni, da gre sesalec za eno polje **naprej v smeri, v katero je obrnjen**.
- `<` pomeni, da se sesalec obrne (za 90 stopinj) na levo.
- `>` pomeni, da se sesalec obrne (za 90 stopinj) na desno.
- `v` ne obstaja več.

Napišite program, ki na osnovi navodila, na primer

```python
pot = "^^<^>^<^^^<^>^>^^^>^"
```

nariše pot. "Slučajno" da tole navodila natanko isto sliko kot prej.

### Rešitev

Ves trik je v tem, kako si zapomniti smer, v katero je obrnjen sesalec. Celoten program je odvisen od tega; enako se nam bo dogajalo v temi, v katero se podajamo v naslednjih predavanjih, objektno programiranje.

Tule sem se odločil tako: imeli bomo spremenljivko `smer`, ki bo imela vrednost 0, ko je sesalec obrnjen desno, 1, ko je obrnjen gor, 2, ko levo in 3, ko desno. Se pravi: 0 je desno, nato pa v pozitivni smeri, smeri nasprotni urinemu kazalcu.

- Prej smo imeli slovar smeri, ki je preslikal premike (`>`, `^`, `<` in `v`) v spremembe koordinat. Zdaj imamo namesto slovarja kar seznam, `smeri = [(1, 0), (0, 1), (-1, 0), (0, -1)]`.
- Koraka `<` in `>` povečata oz. zmanjšata smer za `1` - po modulu 4, tako da se -1 spremeni v 3 in 4 v 0.
