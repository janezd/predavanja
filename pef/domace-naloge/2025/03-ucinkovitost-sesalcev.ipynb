## `sled`

1. Napišite funkcijo `sled(pot, sirina, visina)`, ki vrne seznam koordinat polj, ki jih obišče sesalec na podani poti. 

   Klic `sled(">>^<<<<v^", 5, 5)` vrne `[(0, 0), (1, 0), (2, 0), (2, 1), (1, 1), (0, 1), (0, 0), (0, 1)]`.

   Klic `sled("", 5, 5)` vrne `[(0, 0)]`. Tudi klic `sled("<<<<vvv", 5, 5)` vrne `[(0, 0)]`; neopravljenih premikov ne beležimo.

### Rešitev

To znamo že iz prejšnjih domačih nalog, le polja zlagamo v seznam.

No, vseeno: dva trika. V začetku ne sestavimo praznega seznama, temveč seznam, ki že vsebuje začetno polje. Da je to potrebno, je očitno: tudi če se sesalec sploh ni premaknil, mora funkcija vrniti `[(0, 0)]`.

In ker zadnji element seznama vedno vsebuje trenutni koordinati, ni nobene potrebe, da bi imeli poleg seznama še spremenljivki `x` in `y`.

```python
def sled(pot, sirina, visina):
    obiskano = [(0, 0)]
    for smer in pot:
        dx, dy = {"<": (-1, 0), ">": (1, 0), "^": (0, 1), "v": (0, -1)}[smer]
        nx, ny = obiskano[-1][0] + dx, obiskano[-1][1] + dy
        if 0 <= nx < sirina and 0 <= ny < visina:
            obiskano.append((nx, ny))
    return obiskano
```

## `posesano`

2. Napišite funkcijo `posesano(pot, sirina, visina)`, ki vrne število polj, ki jih je sesalec posesal. (Če isto polje posesa večkrat, ga štejemo samo enkrat.)

    Klic `posesano(">>^<<<<v^", 5, 5)` vrne `6`.

    Klic `posesano("", 5, 5)` vrne `1`. Sesalec se sicer ni premaknil, vendar je posesal vsaj začetno polje.

### Rešitev

Prejšnja funkcija nam vrne seznam. Da preštejemo število elementov brez ponovitev, ga vržemo v množico in vprašamo `len`, koliko elementov vsebuje.

```python
def posesano(pot, sirina, visina):
    return len(set(sled(pot, sirina, visina)))
```

### `posesano2`

3. Napišite funkcijo `posesano2(pot1, pot2, sirina, visina)`, ki vrne število polj, ki jih posesata dva sesalca. Najprej gre na pot prvi, nato drugi. Polja, ki so posesana večkrat, štejemo le enkrat.

    Klic `posesano2("^^>", "^>", 5, 5)` vrne 5; posesana so polja `(0, 0)`, `(0, 1)`, `(0, 2)`, `(1, 1)` in `(1, 2)`.

### Rešitev

Tole pa je, preprosto, število elementov, ki jih vsebuje unija obeh množic.

```python
def posesano2(pot1, pot2, sirina, visina):
    return len(set(sled(pot1, sirina, visina)) | set(sled(pot2, sirina, visina)))
```

Druga možnost je, da seštejemo seznama, ju vržemo v množico in vrnemo število njenih elementov.

```python
def posesano2(pot1, pot2, sirina, visina):
    return len(set(sled(pot1, sirina, visina) + sled(pot2, sirina, visina)))
```

## `posesano_skupaj`

4. Napišite funkcijo `posesano_skupaj(poti, sirina, visina)`, ki je podobna prejšnji funkciji, vendar namesto dve poti prejme cel seznam poti. Tako kot prejšnja funkcija taudi ta vrne skupno število posesanih polj.

    Klic `posesano_skupaj(["^^>", "^>", ">>>"], 5, 5)` vrne `8`.

### Rešitev

Ista finta, le da računamo unijo več množic. V začetku sestavimo prazno množico, nato vanjo dodajamo ("unijamo") množice polj, ki jih posesa posamični sesalec.

```python
def posesano_skupaj(poti, sirina, visina):
    skupaj = set()
    for pot in poti:
        skupaj |= set(sled(pot, sirina, visina))
    return len(skupaj)
```

## `neposesano`

5. Napišite funkcijo `neposesano(poti, sirina, visina)`, ki prejme enake argumente kot prejšnja in vrne število neposesanih polj.

### Rešitev

Tule pa preprosto odštejemo število posesanih polj od celotne površine pravokotnika.

```python
def neposesano(poti, sirina, visina):
    return sirina * visina - posesano_skupaj(poti, sirina, visina)
```

## `prispevki`

6. Napišite funkcijo `prispevki(poti, sirina, visina)`, ki prejme enake argument in vrne seznam s številom polj, ki jih posesa posamezni sesalec (in jih ni posesal noben sesalec pred njim).
 
  Klic `prispevki(["^^>", "^>", ">>>vvvvvv"], 5, 5)` vrne `[4, 1, 3]`, saj prvi sesalec posesa štiri polja, drugi eno polje (ki ga še ni posesal prvi) in tretji tri (premike dol seveda ignoriramo, saj je tam stena).
  
### Rešitev

Tole je - z izjemo v manjši meri prve - edina naloga, pri kateri je bilo potrebno kaj sprogramirati. :)

```python
def prispevki(poti, sirina, visina):
    skupaj = set()
    prispeval = []
    for pot in poti:
        obiskano = set(sled(pot, sirina, visina))
        prispeval.append(len(obiskano - skupaj))
        skupaj |= obiskano
    return prispeval
```

V `skupaj` vzdržujemo množico vseh doslej posesanih polj. Za vsak sesalec bo `obiskano` množica vseh polj, ki jih je obiskal. Novoposesana polja so `obiskano - skupaj` (vsi elementi množice `obiskani`, ki niso v množici `skupaj`); velikost te množice je prispevek sesalca. Prispevek sesalca zabeležimo v seznam `prispeval`, polja, ki jih je posesal, pa "priunijamo" v `skupaj`.
