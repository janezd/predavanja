## Obvezna naloga

Naslednje funkcije morate rešiti z uporabo izpeljanih seznamov, slovarjev oziroma množic. Funkcije torej ne smejo vsebovati drugega kot `return`, v katerem izračunate ali sestavite, kar mora funkcija vrniti.

Argument `pot` podaja seznam polj, ki jih obišče sesalec (v takem vrstnem redu, kot jih je obiskal), na primer [(0, 0), (1, 0), (2, 0), (2, 1)]. Nekatere funkcije sprejemajo še argument `ovire`, ki je množica polj, na katerih stojijo ovire. Sesalec v tej nalogi preprosto prevozi oviro.

Napišite naslednje funkcije.

### 1. `skrajno_gor(pot)`

`skrajno_gor(pot)` vrne največjo y koordinato na poti.

#### Rešitev

Iz seznama `pot` moramo nabrati seznam vseh koordinat `y`, torej `[y for _, y in pot]`. Iščemo največjo vrednost v tem seznamu.

```python
def skrajno_gor(pot):
    return max([y for _, y in pot])
```

### 2. `stolpci(pot)`

`stolpci(pot)` vrne množico x koordinat, ki jih sesalec obišče.

#### Rešitev

To je še lažje: le množico koordinat `x` vrnemo.

```python
def stolpci(pot):
    return {x for x, _ in pot}
```

## 3. `naj_razdalja(pot)`

`naj_razdalja(pot)` vrne razdaljo do najbolj oddaljenega polja. Razdalja do polja (x, y) je |x| + |y| (ne Evklidska, temveč [Manhattanska razdalja](https://en.wikipedia.org/wiki/Taxicab_geometry)!)

#### Rešitev

Sestaviti nam je seznam vseh razdalj, `[x + y for x, y in pot]`. Naloga zahteva največjo vrednost v tem seznamu.

```python
def naj_razdalja(pot):
    return max([x + y for x, y in pot])
```

### 4. `povozene_ovire(pot, ovire)`

`povozene_ovire(pot, ovire)` vrne število povoženih ovir. **Če isto oviro povozi večkrat, jo tudi štejemo večkrat!**

#### Rešitev

Te naloge ne moremo reševati s presekom množic, ker bi na ta način šteli več prehodov čez isto oviro kot en sam prehod.

Če so (x, y) koordinate nekega polja, bo `(x, y) in ovire` enak `True`, če je tam ovira in `False`, če je ni. Ker je `True` toliko kot `1` in `False` toliko kot `0`, moramo le sešteti vrednost `(x, y) in ovire` za vsak `(x, y)` na poti.

```python
def povozene_ovire(pot, ovire):
    return sum([(x, y) in ovire for x, y in pot])
```

Gre tudi malenkost krajše: koordinat nam ni potrebno razpakirati v `x` in `y`:

```python
def povozene_ovire(pot, ovire):
    return sum([polje in ovire for polje in pot])
```

### 5. `zadetki_ovir(pot, ovire)`

`zadetki_ovir(pot, ovire)` vrne slovar, katerega ključi so koordinate ovir, pripadajoče vrednosti pa povedo, kolikokrat je bila posamezna ovira povožena. Klic `zadetki_ovir([(0, 0), (0, 1), (0, 2), (0, 1), (0, 2), (0, 1)], {(0, 0), (0, 1), (4, 2)})` vrne `{(0, 0): 1, (0, 1): 3, (4, 2): 0}`.

#### Rešitev

Gremo čez ovire in za vsako preštejemo, kolikokrat se pojavi v poti.

```python
def zadetki_ovir(pot, ovire):
    return {ovira: pot.count(ovira) for ovira in ovire}
```

Brez izpeljanih slovarjev bi nalogo rešili tako:

```python
def zadetki_ovir(pot, ovire):
    zadetki = dict.fromkeys(ovire, 0)
    for polje in pot:
        if polje in ovire:
            zadetki[polje] += 1
    return zadetki
```

Ta rešitev je (vsaj v teoriji, najbrž pa tudi v praksi) hitrejša. Ona v eni vrstici za vsako oviro prehodi celo pot (znotraj `count`), torej je čas, ki ga potrebuje, sorazmeren produktu števila ovir in dolžine poti. Čas, ki ga potrebuje druga rešitev je sorazmeren dolžini poti (čas, ki ga porabi izraz `polje in ovire`, je neodvisen od števila ovir, saj je `ovire` množica).

Druge rešitve ne moremo (vsaj ne učinkovito in brez velikih sitnosti) spraviti v kaj podobnega izpeljanemu slovarju. Ta, druga rešitev namreč *spreminja* slovar, ki ga sestavlja, izpeljanih slovarjev pa ne moremo spreminjati medtem, ko nastajajo.

### 8. `vodoravni(pot)`

`vodoravni(pot)` vrne število vodoravnih premikov sesalca. Sesalec se vedno premakne bodisi vodoravno bodisi navpično.

#### Rešitev

Opazovati moramo zaporedne koordinate. Konkretno, koordinate `y`: če sta enaki (`y1 == y2`), gre za premik v vodoravni smeri. Tako kot v četrti nalogi spet seštevamo logične vrednosti.

```python
def vodoravni(pot):
    return sum([y1 == y0 for (_, y0), (_, y1) in pairwise(pot)])
```

### 9. `desno(pot)`

`desno(pot)` vrne število premikov na desno.

#### Rešitev

Ista finta.

```python
def desno(pot):
    return sum([x1 > x0 for (x0, _), (x1, _) in pairwise(pot)])
```

### 1. `na_poljih(pot)

`na_poljih(pot)` vrne slovar, katerega ključi so obiskana polja, vrednosti pa povedo, kolikokrat je bilo dotično polje obiskano.

#### Rešitev

Rešitev je lahko takšna:

```python
def na_poljih(pot):
    return {(x, y): pot.count((x, y)) for x, y in pot}
```

Tako kot pri peti nalogi nam tudi tu ni potrebno razpakirati koordinat.

```python
def na_poljih(pot):
    return {polje: pot.count(polje) for polje in pot}
```

Čas, ki ga porabi, je sorazmeren kvadratu dolžine poti. Pri tem posebej boli, da polja, ki se ponovijo večkrat, v slovar dodajamo večkrat in za vsako znova pokličemo `count`. To lahko omilimo tako, da `pot` vržemo v množico in se tako znebimo duplikatov.

```python
def na_poljih(pot):
    return {polje: pot.count(polje) for polje in set(pot)}
```

Ker gre zanka for čez množico, bomo `count` poklicali le še enkrat za vsako obiskano polje. Rešitev, podobna (večvrstični) rešitvi pete naloge pa je še vedno hitrejša.

Mimogrede omenimo, da za takšno štetje obstaja funkcija. Nalogo bi lahko rešili kar takole:

```python
from collections import Counter

def na_poljih(pot):
    return Counter(pot)
```

Ali, še bolj direktno:

```python
from collections import Counter

na_poljih = Counter
```

### 2. `razdalja(pot)`

`razdalja(pot)` vrne skupno razdaljo, ki jo je sesalec prevozil. **Pri tej nalogi se sesalec lahko premakne tudi za več polj**, na primer `[(0, 0), (0, 4), (2, 4), (5, 4), (5, 2)]`. Še vedno pa se premika le vodoravno ali navpično, ne diagonalno.

#### Rešitev

```python
def razdalja(pot):
    return sum([abs(x1 - x0) + abs(y1 - y0) for (x0, y0), (x1, y1) in pairwise(pot)])
```

Ali se sesalec premika vodoravno ali navpično, nas ne briga. Da, v vsakem koraku bo bodisi `x1 - x0` bodisi `y1 - y0` enak 0. Ampak od tega, da prištejemo `0` ne bo nobene škode.
