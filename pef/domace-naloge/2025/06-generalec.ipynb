Za tokratne funkcije morate predpostaviti, da je argument `pot` generator. S klicem `iter(pot)` boste dobili iterator in če na dobljenem iteratorju kličete `next` bo vračala elemente poti. Seveda pa velja tudi `for polje in pot` ali kaj podobnega.

Generatorja/iteratorja, kot vemo, ne moremo, vprašati, koliko elementov bo zgeneriral, pa tudi indeksiranje (`pot[n]`) ne bo delovalo.

Generator lahko generira tudi zeloooooo dolge poti. Če se program zaradi tega noče ustaviti, ga pokončajte z rdečim kvadratkom zgoraj desno.

Argument `ovire` je množica.

Napišite naslednje funkcije.

### 1. zacetek(pot)

`zacetek(pot)` vrne začetno polje poti. To ni nujno `(0, 0)`. Pomislite, kako bi (oz. kako sem) na predavanju dobil prvi element od vas, ko ste bili v vlogi iteratorja.

#### Rešitev

`pot` moramo pripraviti do tega, da nam zgenerira eno polje. `pot[0]` ne bo delovalo, ker generatorja ne moremo indeksirati, `list(pot)` pa sploh ne, ker lahko sestaviti tudi seznam z neskončno elementi. Rešitev je, da pokličemo `iter` (da dobimo iterator) in nato `next).

```python
def zacetek(pot):
    return next(iter(pot))
```

### 2. `nto_polje(pot, n)`

`nto_polj(pot, n)` vrne n-to polje generatorja. Šteti začnemo pri 1.

#### Rešitev

To je podobno, le da moramo izgenerirati `n` elementov.

Lahko bi naredili, recimo, tako:

```python
def nto_polj(pot, n):
    i = iter(pot)
    for _ in range(n - 1):
        next(i)
    return next(i)
```

Lepše je, da zanko kar združimo.

```python
def nto_polje(pot, n):
    for _, polje in zip(range(n), pot):
        pass
    return polje
```

### 3. `zadane(pot, ovire)`

`zadane(pot, ovire)` vrne `True`, če pot vsebuje kakšno oviro.

Potrudite se nalogo rešiti s funkcijo `any`, ki ji podate generator.

#### Rešitev

Tole rešimo enako kot prejšnji teden, le brez oglatih oklepajev: funkciji `any` podamo generator, ki vrača `True` (`False`) za vsa polja, ki vsebujejo (ne vsebujejo) oviro (ovire).

```python
def zadane(pot, ovire):
    return any(polje in ovire for polje in pot)
```

### 4. `zadetki(pot, ovire)`

`zadetki(pot, ovir) vrne generator (to je: fukcija vsebuje bodisi `yield` bodisi `return (... for ... in ...)`), ki vrne vsa polja na poti, v takem vrstnem redu, kot se pojavljajo na poti.

Recimo, da generator generira pot `(0, 0), (1, 0), (0, 0), (1, 0), (0, 0) ...` in tako milijonkrat. Če na polju `(1, 0)` stoji ovira, mora vaš generator pol milijonkrat vrniti `(1, 0)`.

#### Rešitev

```python
def zadetki(pot, ovire):
    return (polje for polje in pot if polje in ovire)
```

### 5. `pot(smeri)` 

`pot(smeri)` prejme smeri (to je, zaporedje `<`, `>`, `^`, `v`) in vrne generator, ki generira pot, ki jo opisujejo smeri. Pot se začne na `(0, 0)`.

Klic `pot(">>^<")` vrne generator, ki generira `(0, 0), (1, 0), (2, 0), (2, 1), (1, 1)`.

`smeri` so generator. Testi bodo (na nek način) preverili, ali vaša funkcija generira neskončno pot v desno, če je pot neskočno zaporedje znakov `>`.

#### Rešitev

Nič takšnega: napišemo funkcijo (oz. generator), ki je podobna funkciji, ki bi sestavila seznam vseh polj na poti, le da namesto, da bi jih dejansko dodajala v seznam, kliče `yield` za vsako polje.

```python
def pot(smeri):
    x = y = 0
    d = {"<": (-1, 0), ">": (1, 0), "v": (0, -1), "^": (0, 1)}
    yield x, y
    for smer in smeri:
        dx, dy = d[smer]
        x += dx
        y += dy
        yield x, y
```

Na začetek smo dali `yield`, da vrnemo tudi prvo polje.
