## Sodobnejši sesalci

Model S1 -- prvi model sesalca, ki ga je izdelalo podjetje **Sesalci Matej** -- je sicer še vedno priljubljen, vendar že dokaj star. S1 je znal le sesati. 

- Na osnovi modela S1 so izdelali tri različne modele: `{"S1A", "S2", "S3"}`: S1A je imel še funkcijo "turbo" (1.8x hitrejše sesanje), S2 je imel zaslon in več tipk, S3 pa prav tako "turbo", vendar drugačen design.

- Na osnovi S3 so izdelali S3A; ta model ima poleg vsega, kar ima S3 še globinsko sesanje in poseben program za odstranjevanje mačjih dlak.

- S1A je bil slepa ulica.

- Na osnovi S2 so izdelali S2+ (ta poleg vsega, kar ima S2, obvlada še vrnitev na izhodišče in rekurzivno sesanje), S2X (k S2 dodaja samo vrnitev na izhodišče) in S2P8 (med sesanjem predvaja reklamna sporočila).

- Na osnovi S2P8 so izdelali S8, ki je nekoliko dražji, funkcionalno pa ne dodaja ničesar k S2P8.

- Na osnovi S8 so izdelali ... in tako naprej. Vse povezave med modeli in lastnosti so našteti v testih.

**Pri tej domači nalogi lahko pišete rekurzivne ali nerekurzivne funkcije. Nekatere bo lažje pisati tako, nekatere drugače. Če boste funkcijo, ki jo je bolj naravno narediti nerekurzivno, poskusili napisati rekurzivno, bo to lahko zoprno. In obratno; če boste funkcijo, ki zahteva rekurzijo, pisali nerekurzivno, boste morali uporabljati stvari, ki se jih (še?) nismo učili.**

### 1. `stevilo_izpeljank`

Napišite funkcijo `stevilo_izpeljank(model)`, ki vrne število vseh modelov, ki so bili posredno ali neposredno izpeljani iz podanega modela.

- Klic `stevilo_izpeljank("S2P8")` vrne 3, ker je bil iz S2P8 izpeljan S8, iz njega pa S8SE in S8PRO.
- Klic `stevilo_izpeljank("S1A")` vrne 0, saj na osnovi tega modela ni nastal noben drug model.

#### Rešitev

Tole spominja na velikost rodbine, oziroma, točneje, število potomcev.

Število izpeljank modela je enako številu direktno izpeljanih modelov ("otrok") in vsoti števila vseh modelov, ki so izpeljanih iz teh, neposredno izpeljanih modelov.

```python
def stevilo_izpeljank(model):
    izpeljank = len(modeli[model])
    for m in modeli[model]:
        izpeljank += stevilo_izpeljank(m)
    return izpeljank
```

Ali pa tako:

```python
def stevilo_izpeljank(model):
    izpeljank = 0
    for m in modeli[model]:
        izpeljank += 1 + stevilo_izpeljank(m)
    return izpeljank
```

Krajše pa

```python
def stevilo_izpeljank(model):
    return len(modeli[model]) + sum(stevilo_izpeljank(m) for m in modeli[model])
```

Oziroma

```python
def stevilo_izpeljank(model):
    return sum(1 + stevilo_izpeljank(m) for m in modeli[model])
```

Kdor pozna `map` (o katerem se sicer nismo učili) pa napiše

```python
def stevilo_izpeljank(model):
    return len(modeli[model]) + sum(map(stevilo_izpeljank, modeli[model]))
```

In, končno, kdor poškili malo naprej ter napiše rešitev, ki le pokliče funkcijo iz naslednje naloge in vrne velikost dobljene množice.

```python
def stevilo_izpeljank(model):
    return len(izpeljanke(model))
```

### 2. `izpeljanke(model)`

Napišite funkcijo `izpeljanke(model)`, ki vrne množico vseh modelov, ki so bili posredno ali neposredno izpeljani iz podanega modela.

- Klic `izpeljanke("S2P8")` vrne `{"S8", "S8SE", "S8PRO"}`.
- Klic `izpeljanke("S1A")` vrne prazno množico, saj na osnovi tega modela ni nastal noben drug model.

#### Rešitev

To je pravzaprav enako, le da namesto števila izpeljank vrnemo množico izpeljank. Namesto `izpeljank = len(modeli[model])` torej pišemo `izp = set(modeli[model])` in namesto seštevanja imamo unijo.

```python
def izpeljanke(model):
    izp = set(modeli[model])
    for m in modeli[model]:
        izp |= izpeljanke(m)
    return izp
```

Obstaja rešitev z generatorji; naj jo razume, kdor jo razume.

```python
def izpeljanke(model):
    return set().union(modeli[model], *(izpeljanke(m) for m in modeli[model]))
```

### 3. `predniki(model)`

Napišite funkcijo `predniki(model)`, ki vrne seznam prednikov modela.

- Klic `predniki("S8SE")` vrne `["S8", "S2P8", "S2", "S1"]`, saj je S8SE izpeljan iz S8, ta pa iz S2P8, ta iz S2, ta pa iz S1.
- Klic `predniki("S1")` vrne prazen seznam, saj S1 nima prednikov.

#### Rešitev

Tega se je boljše lotiti z običajno zanko, brez rekurzije. Za vsak sesalec bo potrebno poiskati prednika; za to si jo morda pametno pripraviti ločeno funkcijo.

```python
def poisci_prejsnjega(model):
    for m, izpeljanke in modeli.items():
        if model in izpeljanke:
            return m
    return None

def predniki(model):
    prej = []
    while True:
        model = poisci_prejsnjega(model)
        if model is None:
            return prej
        prej.append(model)
```

Naloga je torej predvsem naloga iz zank (in slovarjev in funkcij). V `poisci_prejsnjega` je potrebno v `modeli` poiskati ključ, katerega vrednost je množica, ki vsebuje `model`. Ker potrebujemo ključe in množice, uporabimo `modeli.items()`. Znotraj zanke imamo `return` - ta prekine zanko, vsak `break` bi bil tu odveč. Če ne najdemo ničesar, vrnemo `None`.

V drugi funkciji imamo zanko `while`. In to celo neskončno zanko `while`; tako je najbolj praktično. V njej vsakokrat poiščemo prednika trenutnega modela in ga dodamo v seznam. Če je model slučajno `None` - kar pomeni, da smo naleteli na sesalec brez prednikov -, pa vrnemo trenutni seznam.

Če nam neskončne zanke niso všeč, lahko napišemo takole - je pa ta rešitev grša.

```python
def predniki(model):
    prej = [model]
    while prej[-1] != None:
        prej.append(poisci_prejsnjega(prej[-1]))
    return prej[1:-1]
```

`prej[-1]` je vedno enak zadnjemu trenutnemu doslej, torej lahko pišemo tudi `poisci_prejsnjega(prej[-1])`. To bo vrnilo naslednjega prednika, zato ga kar takoj dodamo v seznam `prej.append(poisci_prejsnjega(prej[-1]))`. Da bo to res delovalo, pa moramo v začetku dodati `model`, ki smo ga dobili kot argument. Poleg tega bo na koncu v seznam dodan tudi `None`. Ko se to zgodi, prekinemo zanko (`while prej[-1] != None`). Na koncu vrnemo seznam brez prvega elementa (podani sesalec) in zadnjega elementa (`None`).

To je kar grdo.

V izogib takim grdobijam je Python v neki nedavni verziji dobil "mrožev operator", s katerim lahko prirejamo vrednosti tudi na mestih, kjer je operator `=` prepovedan, recimo znotraj pogoja v `while`.

```python
def predniki(model):
    prej = []
    while (model := poisci_prejsnjega(model)) != None:
        prej.append(model)
    return prej
```

Če vemo, da so imena sesalcev neprazni nizi, lahko tudi izpustimo primerjanje z `None` in dobimo:

```python
def predniki(model):
    prej = []
    while model := poisci_prejsnjega(model):
        prej.append(model)
    return prej
```

Mrožev operator je bil predmet dolge vojne. Meni nikoli ni bil zelo všeč, zato nisem prepričan, ali mi je bolj všeč tale rešitev ali ona, čisto prva.

Če ne pišemo dodatne funkcije, moramo stlačiti zanko `while` v zanko `for`, poleg tega pa v njej ne moremo uporabiti `return`. V Pythonu lahko uporabimo `else` po `for` in dobimo tole ne-nezanimivo rešitev.

```python
def predniki(model):
    prej = []
    while True:
        for m, izpeljanke in modeli.items():
            if model in izpeljanke:
                prej.append(m)
                model = m
                break
        else:
            return prej
```

`return prej` je znotraj `else`, ki je v `for`. Ta se izvede, če se je zanka iztekla do konca (torej: če je ni prekinil `break`). Z drugimi besedami, če za nek model ne najdemo prednika, vrnemo `prej`.

### 4. `lastnosti_modela(model)`

Napišite funkcijo `lastnosti_modela(model)`, ki vrne vse funkcije, ki jih ima model.

- Klic `lastnosti_modela("S8SE")` vrne `{"sesanje", "turbo", "zaslon", "vrnitev na izhodišče", "reklamna sporočila"}`, saj S8SE obvlada vse funkcije, ki jih obvlada S2P8 (to so: sesanje, zaslon, vrnitev na izhodišče in reklamna sporočila), pa tudi turbo, ki ga ima S3A, ki je prav tako izpeljan iz S2P8.

- Klic `lastnosti_modela("S1")` vrne `{"sesanje"}`, saj S1 obvlada samo sesanje.

- Klic `lastnosti_modela("S2+")` vrne `{'sesanje', 'zaslon', 'več tipk', 'rekurzivno sesanje', 'vrnitev na izhodišče'}`. Sesanje je "podedoval" po S1, zaslon in več tipk po S2, vrnitev na izhodišče in rekurzivno sesanje pa dodaja S2+.

#### Rešitev

Vrniti moramo unijo lastnosti tega sesalca in vseh njegovih prednikov.

```python
def lastnosti_modela(model):
    last = lastnosti[model].copy()
    for model in predniki(model):
        last |= lastnosti[model]
    return last
```

Ne spreglejte, da smo naredili kopijo množice - poklicali smo `copy()`! To je pomembno, sicer bi dodajali nove lastnosti v množico `lastnosti[model]`. Ker sem sam reševal to nalogo čisto drugače (glej spodaj), v teste žal nisem vključil preverjanja, ali je slovar z lastnostmi po klicu te funkcije še vedno enak, kot je bil pred njim.

Krajša rešitev je

```python
def lastnosti_modela(model):
    return lastnosti[model].union(*(lastnosti[m] for m in predniki(model)))
```

### 5. `zaostanek(model)`

Napišite funkcijo `zaostanek(model)`, ki pove, kako zaostal je model. Zaostanek modela S2P8 je 2, saj so iz njega izpeljali S8 iz S8 pa S8PRO. Zaostanek modela S2 je 3, saj so iz njega izpeljali, recimo, S2P8, iz tega S8 in iz njega S8SE. (Iz S2 so sicer izpeljali tudi S2+ in iz njega S2++; po tej "liniji" bi bil zaostanek samo 2, vendar funkcija vrne "največji zaostanek".)

#### Rešitev

Tule bo spet praktično uporabiti rekurzivno funkcijo. Rešitev je lahko

```python
def zaostanek(model):
    if not modeli[model]:
        return 0
    zaost = 0
    for m in modeli[model]:
        zaost = max(zaost, zaostanek(m))
    return zaost + 1
```

ali

```python
def zaostanek(model):
    if not modeli[model]:
        return 0
    zaost = 0
    for m in modeli[model]:
        zaost_m = zaostanek(m)
        if zaost_m > zaost:
            zaost = zaost_m
    return zaost + 1
```

ne pa

```python
def zaostanek(model):
    if not modeli[model]:
        return 0
    zaost = 0
    for m in modeli[model]:
        if zaostanek(m) > zaost:
            zaost = zaostanek(m)
    return zaost + 1
```

saj tu, kot smo se učili, stvar eksplodira, ker funkcija večkrat kliče samo sebe z enakimi argumenti.

Za malo krajšo rešitev uporabimo generator:

```python
def zaostanek(model):
    if not modeli[model]:
        return 0
    return max(zaostanek(m) for m in modeli[model]) + 1
```

Če vemo, da je nič neresnična (in nas to ne moti) pa lahko pišemo kar

```python
def zaostanek(model):
    return len(modeli[model]) and max(zaostanek(m) for m in modeli[model]) + 1
```
