Funkcije v tej nalogi morajo biti rekurzivne (čeprav bi se jih seveda dalo napisati tudi brez rekurzije). Torej: nobenih zank, nobenih izpeljanih seznamov, generatorjev ali pomožnih funkcij. Vsaka funkcija obravnava robne pogoje in, če še ni "na robu", rekurzivno kliče samo sebe.

### `1. Vrnitve`

Menda prinaša srečo, če se sesalec vrača na izhodiščno polje. Napišite funkcijo `vrnitve(pot)`, ki vrne število elementov `(0, 0)` v podani poti.

    Klic `vrnitve([])` vrne 0. Klic `vrnitve([(0, 0), (1, 0), (0, 0)])` vrne 2.

#### Rešitev

Tako kot smo kdaj šteli - ali celo vračali - vse lihe elemente seznama, zdaj štejemo vse elemente, ki so enaki `(0, 0)`.

Nekoliko daljša rešitev je

```python
def vrnitve(pot):
    if pot == []:
        return 0
    ostale = vrnitve(pot[1:])
    if pot[0] == (0, 0):
        return 1 + ostale
    return ostale
```

Krajše (a ne nujno lepše) pa tako:

```python
def vrnitve(pot):
    return pot != [] and ((pot[0] == (0, 0)) + vrnitve(pot[1:]))
```

### `2. sodo_diagonal`

Kot je znano, tudi sesalci -- podobno kot črne mačke -- prinašajo nesrečo. Črna mačka prinese nesrečo, če prečka cesto, sesalec pa, če se znajde na diagonali. Vendar: sesalec prinese nesrečo le, če se na diagonali znajde liho-krat.

Napišite funkcijo `sodo_diagonal(pot)`, ki vrne `True`, če `pot` vsebuje sodo število elementov z enakima koordinatama, in `False` sicer.

Klic `sodo_diagonal([(0, 0), (0, 1), (0, 2), (2, 2), (2, 3)])` vrne `True`, saj seznam vsebuje dve (torej: sodo število) diagonalnih polj (namreč: `(0, 0)` in `(2, 2)`). Klic `sodo_diagonal([(0, 0), (0, 1), (0, 2), (2, 2), (2, 3), (3, 3)])` vrne `False`, saj seznam vsebuje tri (torej: liho število) diagonalnih polj (namreč: `(0, 0)`, `(2, 2)` in `(3, 3)`).

Namig: če je trenutno polje diagonalno, bomo vrnili `True`, če je ostanek ...? Če pa trenutno polje ni diagonalno, bomo vrnili `True`, če je ostanek ...?

#### Rešitev

- Če je pot prazna, je število diagonal sodo.
- Če je trenutno polje diagonalno, je skupno število diagonal sodo, če ostanek poti NIMA sodega števila diagonal.
- Če trenutno polje ni diagonalno, je skupno število diagonal sodo, če je skupno število diagonal na ostali poti sodo.

To rešitev zapišemo tako:

```python
def sodo_diagonal(pot):
    if pot == []:
        return True
    if pot[0][0] == pot[0][1]:
        return not sodo_diagonal(pot[1:])
    return sodo_diagonal(pot[1:])
```

Krajše pa tako:

```python
def sodo_diagonal(pot):
    return pot == [] or (pot[0][0] == pot[0][1]) != sodo_diagonal(pot[1:])
```

### `3. trinajst`

Vemo tudi, da je trinajst nesrečna številka, sploh, če se pojavi prevečkrat.

Napišite funkcijo `trinajst(pot, n)`, ki vrne `True`, če se sesalec največ n-krat nahaja v stolpcu 13 (torej: če je v `pot` največ `n` polj, katerih prva koordinata je 13) in `False`, če se v trinajstem stolpcu *nahodi* prevečkrat.

#### Rešitev

Najprej daljša različica.

```python
def trinajst(pot, n):
    if pot == []:
        return True
    if pot[0][0] == 13:
        if n == 0:
            return False
        return trinajst(pot[1:], n - 1)
    return trinajst(pot[1:], n)
```

Če je pot prazna, smo zmagali. Sicer preverimo, ali smo v trinajstem stolpcu. Če je tako in če nam ni dovoljen noben 13. stolpec več, vrnemo `False`. Sicer nadaljujemo z rekurzivnim klicem, vendar dovolimo eno trinajstico manj (zato `n - 1`). Če trenutni stolpec ni 13, pa rekurzivno kličemo funkcijo z enakim število dovoljenih 13. stolpcev (vendar, seveda, brez prvega elementa poti, saj smo tega že preverili).

Nekoliko krajša (in, v bistvu, grša) različica je:

```python
def trinajst(pot, n):
    if pot == []:
        return True
    if pot[0][0] == 13 and n == 0:
        return False
    return trinajst(pot[1:], n - (pot[0][0] == 13))
```

Tu smo združili oba rekurzivna klica v enega in od `n` odšteli `True` (kar je `1`) ali `False` (kar je `0`). To, da tretiramo `boolean` kot `0` ali `1` je grdo in zgledneje narejeni jeziki tega ne pustijo.

Malo drugačen pristop gre vedno do konca poti in na koncu vrne `True`, če `n` ni negativen - se pravi, če nismo "porabili" več 13. stolpcev, kot je dovoljeno.

```python
def trinajst(pot, n):
    if pot == []:
        return n >= 0
    return trinajst(pot[1:], n - (pot[0][0] == 13))
```

Odtod pridemo do še grše - vendar kar zabavne - rešitve:

```python
def trinajst(pot, n):
    return trinajst(pot[1:], n - (pot[0][0] == 13)) if pot else n >= 0
```
